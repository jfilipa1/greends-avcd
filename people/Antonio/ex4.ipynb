```python
import pandas as pd
import zipfile
import random
import seaborn as sns
import matplotlib.pyplot as plt 
df = pd.read_csv('EFIplus_medit.zip',compression='zip', sep=";")
```

```python
mean = []
for i in range(1,10,1):
    sample = pd.Series(random.sample(list(df['temp_ann']), i))
    mean.append(sample.mean())

sns.lineplot(mean)
```

```python
ctab = pd.crosstab(df['temp_ann'], df['Salmo trutta fario'])
print(ctab)
```

```python

sns.boxplot(
    data= df,
    x='temp_ann',
    y='Salmo trutta fario',
    orient="h",
    width=0.4,
    fliersize=1,
    whis= 1,
    color='#319ebc', 
    )
sns.stripplot(
    data= df,
    x='temp_ann',
    y='Salmo trutta fario',
    orient="h",
    jitter=0.2, 
    linewidth=0, 
    color='#9e2a3f', 
    alpha=.2,
    s=2, 
    )
plt.tick_params(left=False)
sns.despine(left=True)
plt.xlabel('Mean Annual Temperature C')
plt.ylabel(' yes fish =1                        no fish= 0')
```
