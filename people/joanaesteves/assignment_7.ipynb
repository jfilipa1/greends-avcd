The dataset in *EFIplus_medit.zip* will be used for this assignment.

```python
import numpy as np # for getting our distribution
import pandas as pd # to handle data frames
import matplotlib.pyplot as plt # for plotting
import seaborn as sns # for plotting
from scipy import stats # to compute statistics
from scipy.cluster.hierarchy import dendrogram # to create dendrogram
from scipy.cluster.hierarchy import linkage
from sklearn.cluster import AgglomerativeClustering # to perform agglomerative clustering
```

1. Run agglomerative cluster analysis using different linkage methods based on quantitative environmental 
variables (see previous exercise), using sites from the Douro and Tejo basins.

```python
df = pd.read_csv('EFIplus_medit.zip',compression='zip', sep=";")
```

**Create subsets (using Tejo and Douro) from our dataframe**

```python
df = df[(df['Catchment_name']=='Douro') | (df['Catchment_name']=='Tejo')]
```

**Choose the columns with environmental variables**

```python
columns = ['Altitude','Actual_river_slope','Elevation_mean_catch','prec_ann_catch','temp_ann','temp_jan','temp_jul']
df = df[columns]
df = df.dropna()
df = df.reset_index(drop=True)
```

```python
# Dendrogram with single linkage
dendrogram_plot = dendrogram(linkage(df, method='single'))
plt.title('single linkage')
plt.xlabel('objects')
plt.ylabel('Distance')

plt.xticks(color='w') 
plt.show()
```

2. Plot a heatmap and a dendrogram by clustering the rows (sites) using average linkage 

```python
sns.clustermap(df, col_cluster=False, row_cluster=True, method='average')
```
